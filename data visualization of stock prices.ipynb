```python
#adding all the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading the aapl csv file
aapl_data = pd.read_csv("data/AAPL.csv")
```

```python
#reading the google csv file
google = pd.read_csv("data/GOOG.csv")
```

```python
years = 14
timeperiod = years * 252

fig = plt.figure(figsize = (13, 8))
ax = fig.add_subplot(1,1, 1)
ax.plot(aapl_data["Date"][-timeperiod:], aapl_data["Adj Close"][-timeperiod:], 'b-', label = "Apple")
ax.plot(google["Date"][-timeperiod:], google["Adj Close"][-timeperiod:], 'g-', label = "Google")
ax.set_xticks(np.arange(0, len(aapl_data[-timeperiod:]), 126))
plt.xticks(rotation=90)

# add legend
ax.legend(loc = "best")
ax.set_xlabel("Date")
ax.set_ylabel("Adjusted Closing Price")
ax.set_title("Comparison of Stock Adjusted Closing Price for Apple and Google over the last " + str(years) + " years.")
# end

#start the arrow
ax.annotate("Start of the Great Recession", xy=('2008-07-14', 350), 
             xytext = ('2008-07-14', 1000),
             arrowprops={
                 "facecolor": "red",
                 "headwidth": 20,
                 "headlength": 10,
                 "width": 10},
            horizontalalignment= "left",
            verticalalignment="top"
             )
# end arrow

plt.show()
```
